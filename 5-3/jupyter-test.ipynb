```python
import pandas as pd
import tensorflow as tf

# a+b 연산
a = tf.constant(100)
b = tf.constant(50)
add_op = a + b

# 변수 v 선언하기
v = tf.Variable(0)

# 변수 v에 add_op의 결과 대입하기
let_op = tf.assign(v, add_op)

# 세션 시작하기
sess = tf.Session()

# 변수 초기화하기
sess.run(tf.global_variables_initializer())

# let_op 시작하기
sess.run(let_op)

# 출력하기
print(sess.run(v))
```
